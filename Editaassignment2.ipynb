```python
ta.columns
ta = ta.drop(['Unnamed: 40', 'Unnamed: 41','Unnamed: 42', 'Unnamed: 43', 'Unnamed: 44', 'Unnamed: 45','Unnamed: 46','Unnamed: 47', 'Unnamed: 48', 'Unnamed: 49','Unnamed: 50'], axis=1)
ta = ta.drop(['Range','Volat = stdev of LN' ], axis=1)
ta.columns = ta.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
ta.describe()
```

```python
plt.title("histogram of price", fontsize=18) 
plt.hist(ta['price'], density=True, bins=50, color="Blue", edgecolor='Red')
plt.grid(True)
```

Most common price averaged between 20.5-21 and only a few went above 22.

```python
plt.title("histogram of trade volume", fontsize=18) 
plt.hist(ta['volume'], density=True, bins=50, color="black", edgecolor='pink')
plt.grid(True)
```

trade volume ranged between none to 40, only a few individuals went above that number

```python
pp=ta[['price', 'offer', 'bid', 'user_buy', 'user_sell', 'net']]

sns.pairplot(pp, height=2)
```

```python
sns.swarmplot(x="price", y="volume", data=ta)
plt.title('price verus volume swarmplot',fontsize=16)

plt.show()
```

most trade volume occured on prices of 20-21 or below
